```python
#check unique values in Furnishing column 
data['Furnishing'].unique()
```

```python
#filters the data DataFrame to select only rows where the Furnishing is either 'semi_furnished' or 'Semi'
data = data[(data['Furnishing'] == 'semi_furnished') | (data['Furnishing'] == 'Semi')]
data.head()
```

```python
#make it clean and easy to access and readable
data.columns = data.columns.str.lower()
data.columns= data.columns.str.replace(' ','_')
data.columns= data.columns.str.replace('#','num')
data.head()
```

```python
data['bhk'] = data['num_of_rooms'].str.extract(r'(\d+)').astype(int).astype(str) + " BHK"
```

```python
data.info()
data.describe()
```

```python
data['price_per_sqft'] = data['total_price']/data['super_area']
data.head()
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:Him2002@localhost/pricing_engine")
```

```python
#load the data into sql server
try:
    data.to_sql(name='data', con=engine, index=False, if_exists = 'append')
    print("Import Successfully")
except Exception as e:
    print(f'There is an error : {e}')
```
