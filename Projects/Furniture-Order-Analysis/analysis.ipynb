```python
#read data from file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#rename columns
#df.rename(columns={'Order Id': 'order id', 'City':'city'})
df.columns = df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new columns discount, sales price and profits
df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sales_price'] = df['list_price'] - df['discount']
df['profit'] = df['sales_price'] - df['cost_price']
df.head(5)
```

```python
#convert order date from object datatype to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format= '%Y-%m-%d' )
```

```python
#drop cost price, list price, discount_percent coulmns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace = True)
```
